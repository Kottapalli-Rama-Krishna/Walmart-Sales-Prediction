```python
# Outliers - Detection of outliers using boxplot
for col in col_names:
    if walmart[col].dtype!="object":
        plt.boxplot(walmart[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
walmart.info()
```

```python
# Columns where existing outliers need to be treated
col_list = ['Weekly_Sales','Holiday_Flag','Temperature','Unemployment']
```

```python
# Using IQR finding the lower and upper whisker values and considering the data which is inside the whiskers boundaries
for i in col_list:
    Q1=walmart[i].quantile(0.25)
    Q2=walmart[i].quantile(0.50)
    Q3=walmart[i].quantile(0.75)
    IQR=Q3-Q1
    lower_range =Q1-1.5*IQR
    upper_range =Q3+1.5*IQR
    walmart = walmart[(walmart[i]>=lower_range) & (walmart[i]<=upper_range)]
```

```python
# After removing the data beyond the whiskers plotting the boxplot again
for col in col_names:
    if walmart[col].dtype!="object":
        plt.boxplot(walmart[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
# Since there are still outliers present in Unemployment, running the same step again 
for i in col_list:
    Q1=walmart[i].quantile(0.25)
    Q2=walmart[i].quantile(0.50)
    Q3=walmart[i].quantile(0.75)
    IQR=Q3-Q1
    lower_range =Q1-1.5*IQR
    upper_range =Q3+1.5*IQR
    walmart = walmart[(walmart[i]>=lower_range) & (walmart[i]<=upper_range)]
```

```python
for col in col_names:
    if walmart[col].dtype!="object":
        plt.boxplot(walmart[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

Now all the features are free from outliers and ready for the feature extraction

# Variance Inflation Factor 

Variance inflation factor (VIF) is a feature extraction method which is used to select features from the data for building the ML model by removing the interdependent features from the data. 
If the VIF value is below 5 or 6 or 7 depending on the dataset, those features are considered for the model building.
So only independent columns and non-object columns are considered for performing VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in walmart.columns:
    if((walmart[col].dtype!='object') & (col!='Weekly_Sales')):
        col_list.append(col)
```

```python
col_list
```

```python
X = walmart[col_list] #numerical data
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF values'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))] 
vif_data
```

# Model building

```python
walmart_new = walmart[['Store','CPI']]
walmart_new
```

```python
x = walmart_new # Independent variables for model building
y = walmart['Weekly_Sales'] # Target variable for future prediction
```

# Splitting the data into train data and test data for training our model

```python
# Here we are taking 70% of our data to train the model and 30% of our data to test the built model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.7, random_state=20)
```
